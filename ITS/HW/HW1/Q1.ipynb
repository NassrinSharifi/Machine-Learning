```python
import random
import matplotlib.pyplot as plt


def take_step(location: int) -> int:
    probability = random.random()
    if probability <= 0.8:
        step = 1
    else:
        step = -1
    location += step

    # to avoid negative locations 
    if location < 0:
        location = 1
    return location


```

### Part 1

```python
random.seed(42)
current_location = 0
locations = [current_location]
for _ in range(100):
    current_location = take_step(current_location)
    locations.append(current_location)
current_location

```

```python
def plot(locations: list):
    plt.plot(range(1, len(locations) + 1), locations, c="green")
    plt.xlabel("Step number")
    plt.ylabel("Location")
    plt.title("Movement")
    plt.savefig("Movement")
plot(locations)
```

### Part 2

```python
random.seed(42)
i = 0
current_location = 0
while current_location < 100:
    i += 1
    current_location = take_step(current_location)
i
```

### Part 3

```python
to_100_steps = []
n_of_trial = 1000
for j in range(n_of_trial):
    i = 0
    current_location = 0
    while current_location < 100:
        i += 1
        current_location = take_step(current_location)
    to_100_steps.append(i)

plt.hist(to_100_steps, density=True)
plt.xlabel("Steps")
plt.ylabel("Probability")
plt.savefig("part3")
plt.show()
```

### Part 4

```python
def take_step_V2(location: int) -> list:
    probability = random.random()
    if probability <= 0.8:
        step = 1
    else:
        step = -1

    location += step

    # to avoid negative locations 
    if location < 0:
        location = 1
    return [location, step]


locations = [0]
steps = []
to_100_steps = []
n_of_trial = 1000
for j in range(n_of_trial):
    i = 0
    current_location = 0
    while current_location < 100:
        i += 1
        current_location, step = take_step_V2(current_location)
        locations.append(current_location)
        steps.append(step)
        if len(steps) >= 2:
            if steps[-1] == 1 and steps[-2] == 1:
                step = 2
                steps.append(step)
                current_location += step
                locations.append(current_location)
                i += 1
    to_100_steps.append(i)

locations
```
